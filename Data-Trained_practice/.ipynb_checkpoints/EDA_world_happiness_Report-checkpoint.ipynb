### Wolrd Happiness Report

The World Happiness Report is a landmark survey of the state of global happiness. The first report was published in 2012, the second in 2013, the third in 2015, and the fourth in the 2016 Update. The World Happiness 2017, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th. The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

```python
# let's divide the ranks into quartile
quartile=['Forth','Third','Second','First']
ds['Quartile'] = pd.qcut(ds['Happiness Score'],q=4,labels=quartile)
ds.head()
```

```python
# let's see how the columns are related to the each other
palette = {'tab:green','tab:blue','tab:orange','tab:red'}


sns.pairplot(ds , hue ='Quartile',palette=palette)
plt.show()
```

## Obsarvation :-
From above paire plot we can find below facts.

Happiness score is possitivly correlated with some fectors like Health, Freedom, Famility and Economy, Trust on government. indicating that happy contries having more trust on govt, having good familiy and health. 
country havnig high GDP also makes people happy living in those country.


```python
corr_max = ds.iloc[ :, 1:].corr()
corr_max
plt.figure(figsize=(12,10))
sns.heatmap(corr_max,annot = True,cmap='cool',linewidth=1,linecolor='White')
plt.show()
```

## Obsarvation:-

Generosity is comparativly less correlated with Happiness Score, while it is a key fector in world happiness report.

### Compare different happiness variable of any country with top 10 country , i am taking india as one of the contry, but using below code we can take any country.

```python

df = ds.set_index('Country')
df.drop('Region',axis=1, inplace=True)
ds_top10_to_india = df.head(10).append(df.loc[(df.index=='India')])   # Change country name here 
ds_top10_to_india.drop(columns=['Happiness Rank','Quartile'], inplace=True)
```

```python
ds_top10_to_india
```

```python
for a in ds_top10_to_india.columns:
    plt.figure(figsize=(15,5))
    sns.barplot(x=ds_top10_to_india.index,y=ds_top10_to_india[a].values)
    plt.title(a)
    plt.show()
```
